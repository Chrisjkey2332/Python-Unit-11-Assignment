```python
from random import randint

possibleChoice = ["rock", "paper", "scissors"]

class player:
    def __init__(self):
        self.choice = "None"
        self.score = 0
    def get_input(self, choice, run):
        z = 0
        for x in range(3):
            if choice == possibleChoice[x]:
                z = 1
        if z == 0:
            print("Thank you for playing.")
            run = False
            return run
        if z == 1:
            print(f"You have chosen your champion! {choice}")
            self.choice = choice
            return run
    def return_choice(self):
        return self.choice
    def check(self, computerPick):
        if self.choice != "None":
            if self.choice == "rock" and computerPick == "scissors":
                print("Awesome! You won!")
                self.score = self.score + 1
            elif self.choice == "rock" and computerPick == "paper":
                print("Sucks to suck.")
                self.score = self.score - 1
            if self.choice == "paper" and computerPick == "rock":
                print("Awesome! You won!")
                self.score = self.score + 1
            elif self.choice == "paper" and computerPick == "scissors":
                print("Sucks to suck.")
                self.score = self.score - 1
            if self.choice == "scissors" and computerPick == "paper":
                print("Awesome! You won!")
                self.score = self.score + 1
            elif self.choice == "scissors" and computerPick == "rock":
                print("Sucks to suck.")
                self.score = self.score - 1
            
    def return_score(self):
        return self.score

player = player()
run = True
while run:
    run = player.get_input(input("Pick your champion! (Rock, Paper, or Scissors) "), run)
    if run == True:
        computerPick  = possibleChoice[randint(0, 2)]
        print(f"Bot chose his Champion! {computerPick}")
        player.check(computerPick)
    print(player.return_score())
```
